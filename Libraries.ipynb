# NumPy Section

**Q1.** Create a 1D NumPy array of 15 random integers between 10 and 50. Display all elements greater than the mean value of the array.

```python

import numpy as np
import pandas as pd
arr = np.random.randint(10, 51, 15)
mean_val = np.mean(arr)
result = arr[arr > mean_val]
result
```

**Q2.** Generate a 4x4 NumPy matrix with values from 1 to 16. Replace all even numbers with 0 using array indexing (without using loops).

```python
import numpy as np
matrix = np.arange(1, 17).reshape(4, 4)
matrix[matrix % 2 == 0] = 0
matrix
```

**Q3.** Create two arrays `A` and `B` of size (3,3) filled with random integers (0–10). Compute their element-wise product and matrix multiplication.

```python
import numpy as np
A = np.random.randint(0, 11, (3, 3))
B = np.random.randint(0, 11, (3, 3))
element_wise = A * B
matrix_mul = A @ B
element_wise, matrix_mul
```

**Q4.** Create a 1D array of 20 evenly spaced numbers between 0 and 10. Reshape it into a 4x5 matrix and extract the 2nd column using slicing.

```python
import numpy as np
arr = np.linspace(0, 10, 20).reshape(4, 5)
column_2 = arr[:, 1]
column_2
```

# Matplotlib Section

**Q5.** Plot a line graph for `y = x²` for x in range(-10, 11). Label axes, add a title, and show grid lines.

```python
import matplotlib.pyplot as plt
import numpy as np

x = np.arange(-10, 11)
y = x**2
plt.plot(x, y)
plt.xlabel('x')
plt.ylabel('y = x²')
plt.title('Line Graph of y = x²')
plt.grid(True)
plt.show()
```

**Q6.** Draw a bar chart showing the population of 5 cities of your choice. Label the axes and give a title.

```python
cities = ['Karachi', 'Lahore', 'Islamabad', 'Quetta', 'Peshawar']
population = [16000000, 11000000, 1200000, 1000000, 2000000]

plt.bar(cities, population)
plt.xlabel('Cities')
plt.ylabel('Population')
plt.title('Population of 5 Cities')
plt.show()
```

**Q7.** Generate a scatter plot of 50 random x and y values (0–100). Use a different color and marker style.

```python
x = np.random.randint(0, 101, 50)
y = np.random.randint(0, 101, 50)

plt.scatter(x, y, color='red', marker='x')
plt.xlabel('X Values')
plt.ylabel('Y Values')
plt.title('Scatter Plot of Random Values')
plt.show()
```
